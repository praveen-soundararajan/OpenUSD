```python
from pxr import Usd

# Create a new USD stage with root layer named "prims.usda":
stage: Usd.Stage = Usd.Stage.CreateNew("assets/prims.usda")

# Define a typeless prim
stage.DefinePrim("/hello")

# Define a typed prim: Sphere
stage.DefinePrim("/world", "Sphere")

stage.Save()
```

```python
from pxr import Usd, UsdGeom

file_path = "assets/sphere_prim.usda"
stage: Usd.Stage = Usd.Stage.CreateNew(file_path)

# Define a Sphere prim at /hello
sphere: UsdGeom.Sphere = UsdGeom.Sphere.Define(stage, "/hello")
sphere.CreateRadiusAttr().Set(2)

stage.Save()

print(stage.ExportToString())
```

```python
from pxr import Usd, UsdGeom

file_path = "assets/prim_hierarchy.usda"
stage: Usd.Stage = Usd.Stage.CreateNew(file_path)

# Define a Scope prim at /Geometry
geom_scope: UsdGeom.Scope = UsdGeom.Scope.Define(stage, "/Geometry")

# Define an Xform prim as a child of /Geometry
xform: UsdGeom.Xform = UsdGeom.Xform.Define(stage, geom_scope.GetPath().AppendPath("GroupTransform"))

# Define a cube inside the Xform
cube: UsdGeom.Cube = UsdGeom.Cube.Define(stage, xform.GetPath().AppendPath("Box"))

stage.Save()

print(stage.ExportToString())
```

```python
from pxr import Usd, UsdGeom

layer_path = "assets/second_stage.usda"
stage = Usd.Stage.Open(layer_path)

# Get the Cube prim at /Geometry/GroupTransform/Box
cube = UsdGeom.Cube(stage.GetPrimAtPath("/Geometry/GroupTransform/Box"))

# List all attribute names on Cube
cube_attrs = cube.GetSchemaAttributeNames()
for attr in cube_attrs:
    print(attr)
```

```python
#usda 1.0
(
    defaultPrim = "Geometry"
)

def Scope "Geometry"
{
    def Xform "GroupTransform"
    {
        def Cube "Box"
        {
            double size = 2
            color3f[] displayColor = [(0.0, 0.0, 1.0)]  # initial cube color (blue)
        }
    }
}
```

```python
from pxr import Usd, UsdGeom

layer_path = "assets/second_stage.usda"
stage = Usd.Stage.Open(layer_path)

cube = UsdGeom.Cube(stage.GetPrimAtPath("/Geometry/GroupTransform/Box"))

cube_color_attr = cube.GetDisplayColorAttr()
cube_color_attr.Set([(1.0, 0.0, 0.0)])  # set color to RED

stage.Save()
print("Updated cube color successfully.")
```
